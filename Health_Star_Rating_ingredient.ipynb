```python
from keras.preprocessing.text import text_to_word_sequence
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import string
```

```python
def HSR(name, ingredients):
  rating = 0.0
  n = 0
  out = []
  result = text_to_word_sequence(ingredients, filters='!"$%&()*+,-./:;<=>?[\\]^_`{|}~\t\n')
  for a in Eggs:
    if a in result:
      rating = rating + 3.9
      n = n + 1
      break
  for b in FVNL:
    if b in result:
      rating = rating + 3.7
      n = n + 1
      break
  for c in Seafood:
    if c in result:
      rating = rating + 3.7
      n = n + 1
      break
  for d in Cereal:
    if d in result:
      rating = rating + 3.2
      n = n + 1
      break
  for e in Oil:
    if e in result:
      rating = rating + 3.0
      n = n + 1
      break
  for f in Dairy:
    if f in result:
      rating = rating + 2.9
      n = n + 1
      break
  for g in NABeverage:
    if g in result:
      rating = rating + 2.5
      n = n + 1
      break
  for h in Sauce:
    if h in result:
      rating = rating + 2.5
      n = n + 1
      break
  for i in Meat:
    if i in result:
      rating = rating + 2.2
      n = n + 1
      break
  for j in Bread:
    if j in result:
      rating = rating + 2.1
      n = n + 1
      break
  for k in Sugar:
    if k in result:
      rating = rating + 1.5
      n = n + 1
      break
  for l in Confectionery:
    if l in result:
      rating = rating + 1.1
      n = n + 1
      break
  f_rating = rating/n
  if 0.0 < f_rating <= 2.0:
    value = "Low"
  elif 2.001 < f_rating <= 3.0:
    value = "Medium"
  elif f_rating > 3.0:
    value = "High"
  out.append(name)
  out.append("{:.2f}".format(f_rating))
  out.append(value)
  return out
```
