```
df_train = pd.read_csv('/content/gdrive/MyDrive/MAJOR/datasets/Ingredients dataset/ingredients v1 - Copy.csv')
df_train.head()
```

```
all_ingredients = set()

for ingredients in df_train['features.value']:
   ing=[]
   list_of_ingredients =str(ingredients)
   ing=list_of_ingredients.split(',')
   all_ingredients = all_ingredients | set(ing)
```

```
vocabulary = {}
for ingredient, i in zip(all_ingredients, range(len(all_ingredients))):
    vocabulary[ingredient] = i
```
